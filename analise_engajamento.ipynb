# An√°lise Univariada

```python
# Fun√ß√£o para an√°lise de outliers com base em IQR para vari√°veis num√©ricas com distribui√ß√£o assim√©trica
def analisar_outliers(df, coluna):
    Q1 = df[coluna].quantile(0.25)
    Q3 = df[coluna].quantile(0.75)
    IQR = Q3 - Q1
    limite_inferior = Q1 - 1.5 * IQR
    limite_superior = Q3 + 1.5 * IQR
    print(f"Q1: {Q1}")
    print(f"Q3: {Q3}")
    print(f"IQR: {IQR}")
    print(f"Limite Inferior: {limite_inferior}")
    print(f"Limite Superior: {limite_superior}")
```

```python
# Cria√ß√£o de df com duplicatas e df sem duplicatas para tratamento
df_duplicados = df[df.duplicated(subset='shortcode', keep=False)]
df_unicos = df[~df.duplicated(subset='shortcode', keep=False)]
```

```python
# Verifica√ß√£o das varia√ß√µes nos valores das duplicatas
variacao_duplicados = df_duplicados.groupby('shortcode').agg(
    min_comments=('comments', 'min'),
    max_comments=('comments', 'max'),
    min_likes=('likes', 'min'),
    max_likes=('likes', 'max'),
    count=('shortcode', 'size')  # N√∫mero de duplicatas
).reset_index()

# Calcular as varia√ß√µes
variacao_duplicados['diff_comments'] = variacao_duplicados['max_comments'] - variacao_duplicados['min_comments']
variacao_duplicados['diff_likes'] = variacao_duplicados['max_likes'] - variacao_duplicados['min_likes']

# Filtrar apenas os casos onde h√° varia√ß√£o
variacao_com_variacao = variacao_duplicados[(variacao_duplicados['diff_comments'] > 0) | (variacao_duplicados['diff_likes'] > 0)]

# Exibir o resultado
print(variacao_com_variacao)
```

O resultado mostra que h√° varia√ß√£o nos valores de coment√°rios e likes dentros das duplicatas

### Necessidade de tratamento de duplicatas baseadas em shortcode:

Como o √∫nico elemento que varia entre os registros de duplicatas √© a quantidade de coment√°rios ou de likes, devemos selecionar o com a maior quantidade de likes ou coment√°rios dentre as duplicatas

```python
# Selecionar o registro com mais likes e, em caso de empate, mais comments
df_tratado = df_duplicados.groupby('shortcode').apply(
    lambda x: x.loc[x['likes'].idxmax()] if x['likes'].max() > 0 else x.loc[x['comments'].idxmax()]
).reset_index(drop=True)

# Exibir o DataFrame tratado
df_tratado
```

```python
df = pd.concat([df_unicos, df_tratado], ignore_index=True) # Concatenando dfs criando vers√£o sem duplicatas
df.duplicated(subset='shortcode').sum() # Verificando se ainda h√° duplicatas
```

Agora todos os registros cont√©m apenas um √∫nico registro no que tange a publica√ß√£o

```python
# Fun√ß√£o de contagem de caracteres das legendas que ignora 0
def conta_caracteres(texto):
    if isinstance(texto, str):
        return len(texto)
    else:
        return 0

df['caption_length'] = df['caption'].apply(conta_caracteres) # Cria√ß√£o de coluna com tamanho da legenda
```

Al√©m disso, a an√°lise de Spearman indica uma correla√ß√£o significativa entre:

- Seguidores e engajaento total: h√°, al√©m de uma correla√ß√£o linear, uma correla√ß√£o em que os valores de engajamento podem ser explicados pela vari√¢ncia n√£o necessariamente linear e constante dos valores da quantidade de seguidores.



### An√°lise da correla√ß√£o entre `followers` e `engagement`

```python
from scipy import stats

# C√°lculo do coeficiente de correla√ß√£o de Pearson
correlation_coef, p_value = stats.pearsonr(df['followers'], df['engagement'])

print(f'Coeficiente de Correla√ß√£o de Pearson: {correlation_coef:.2f}, p-valor: {p_value:.4f}')
print(f'Signific√¢ncia: {"Sim" if p_value < 0.05 else "N√£o"}')
print(f'R¬≤: {correlation_coef**2:.2f}')

# Ajuste da reta de regress√£o linear
slope, intercept, r_value, p_value, std_err = stats.linregress(df['followers'], df['engagement'])
line = slope * df['followers'] + intercept

# Criando o scatter plot
plt.figure(figsize=(12, 8))
plt.scatter(df['followers'], df['engagement'], alpha=0.6, color='blue', edgecolors='w', s=100)

# Adicionando a reta de regress√£o
plt.plot(df['followers'], line, color='red', linewidth=2, label=f'Regress√£o Linear\n$y={slope:.2f}x + {intercept:.2f}$\n$r={correlation_coef:.2f}$')

# Adicionando detalhes ao gr√°fico
plt.title('Rela√ß√£o entre Seguidores e Engajamento', fontsize=16)
plt.xlabel('N√∫mero de Seguidores', fontsize=14)
plt.ylabel('Engajamento', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(fontsize=12)
plt.tight_layout()

# Mostrando o gr√°fico
plt.show()
```

√â possivel estabelecer que:
- A an√°lise revela uma rela√ß√£o significativa entre a quantidade de seguidores e o engajamento, comprovada pelo p-valor extremamente baixo, indicando que a correla√ß√£o observada n√£o √© aleat√≥ria.

- 62% da vari√¢ncia nos valores de engajamento √© explicada pela varia√ß√£o na quantidade de seguidores, o que demonstra uma conex√£o substancial entre essas vari√°veis.

- Existe uma correla√ß√£o positiva: √† medida que o n√∫mero de seguidores aumenta, o engajamento tamb√©m tende a aumentar; da mesma forma, um n√∫mero menor de seguidores geralmente est√° associado a um engajamento menor.

Essa tend√™ncia n√£o √© perfeitamente alinhada (linear), mas existe um padr√£o de crescimento que Spearman mostra melhor do que Pearson.

- No 1% superior dos dados (Spearman = 0.42 e Pearson = 0.81): Entre os maiores perfis, a rela√ß√£o parece ser mais direta e linear, o que explica o alto coeficiente de Pearson (0.81). Isso significa que, nesse pequeno grupo, perfis com mais seguidores tendem a ter engajamento proporcionalmente mais alto. No entanto, o coeficiente de Spearman √© mais baixo (0.42), o que indica que a ordem ou o padr√£o de crescimento de engajamento em rela√ß√£o aos seguidores varia mais.
```python
# Teste da normalidade das distribui√ß√µes

shapiro_fotos = stats.shapiro(df_fotos['engagement'])
shapiro_videos = stats.shapiro(df_videos['engagement'])

print("Shapiro-Wilk Test para Fotos: Estat√≠stica =", shapiro_fotos.statistic, ", p-valor =", shapiro_fotos.pvalue)
print("Shapiro-Wilk Test para V√≠deos: Estat√≠stica =", shapiro_videos.statistic, ", p-valor =", shapiro_videos.pvalue)

# Teste Kolmogorov-Smirnov
ks_fotos = stats.kstest(df_fotos['engagement'], 'norm')
ks_videos = stats.kstest(df_videos['engagement'], 'norm')

print("Kolmogorov-Smirnov Test para Fotos: Estat√≠stica =", ks_fotos.statistic, ", p-valor =", ks_fotos.pvalue)
print("Kolmogorov-Smirnov Test para V√≠deos: Estat√≠stica =", ks_videos.statistic, ", p-valor =", ks_videos.pvalue)

# Teste Anderson-Darling
anderson_fotos = stats.anderson(df_fotos['engagement'], dist='norm')
anderson_videos = stats.anderson(df_videos['engagement'], dist='norm')

print("Anderson-Darling Test para Fotos: Estat√≠stica =", anderson_fotos.statistic, ", p-valor =", anderson_fotos.critical_values)
print("Anderson-Darling Test para V√≠deos: Estat√≠stica =", anderson_videos.statistic, ", p-valor =", anderson_videos.critical_values)
```

- Os resultados de teste de normalidade garantem que a distribui√ß√£o de fotos e de v√≠deos n√£o possuem uma distribui√ß√£o normal, assim sendo o melhor m√©todo para an√°lise da vari√¢ncia ser√° o de Mann-Whitney

```python
# Teste de Mann-Whitney para avaliar real signific√¢ncia da diferen√ßa entre valores de engajamento para cada classe (foto/v√≠deo)
# Verifica√ß√£o se a associa√ß√£o n√£o √© casual
statistics, p_value = stats.mannwhitneyu(df_fotos['engagement'], df_videos['engagement'], alternative='two-sided')

print("Teste de Mann-Whitney: Estat√≠stica =", statistics, ", p-valor =", p_value)
```

```python
# Teste com dados limpos de outliers
statistics, p_value = stats.mannwhitneyu(df_fotos_sem_outliers['engagement'], df_videos_sem_outliers['engagement'], alternative='two-sided')

print("Teste de Mann-Whitney: Estat√≠stica =", statistics, ", p-valor =", p_value)
```

- O teste de Mann-Whitney foi utilizado para verificar a signific√¢ncia da associa√ß√£o entre o tipo de post (v√≠deo/foto) e o n√≠vel de engajamento, buscando determinar se essa associa√ß√£o realmente existe ou se √© algo aleat√≥rio na amostra.
- Esse teste foi escolhido porque atende √†s condi√ß√µes da an√°lise: as distribui√ß√µes dos dois grupos s√£o semelhantes (ambos assim√©tricos), as classes s√£o independentes e h√° uma vari√°vel qualitativa (tipo de post) e uma quantitativa (engajamento).
- Com os dados completos, o teste revelou, pelo p-valor >= 0,05, que n√£o h√° evid√™ncias significativas de uma associa√ß√£o entre o tipo de post e o engajamento. Isso sugere que a diferen√ßa observada pode ser fruto do acaso.
- No entanto, ao aplicar o teste ap√≥s a remo√ß√£o dos outliers (valores acima do IQR m√°ximo), o p-valor < 0,05 indica que h√° uma associa√ß√£o significativa entre o tipo de post e o engajamento.
- Concluindo, √© poss√≠vel afirmar que, dentro da faixa t√≠pica de valores de engajamento (sem outliers), existe uma diferen√ßa significativa de engajamento entre v√≠deos e fotos, indicando uma associa√ß√£o real entre o tipo de postagem e a quantidade de engajamento.
